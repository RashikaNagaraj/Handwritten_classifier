```python
count_correct_prediction=0  
count_total_prediction=0
for i in range(len(x_test)):
    d=x_test[i]
    q = numpy.zeros(26)
    for j in range(784):
        if(d[j]==0):
            for k in range(26):
                q[k]+=math.log(f0[k][j])
        else:
            for k in range(26):
                q[k]+=math.log(f1[k][j])
                
    for l in range(26):
        q[l]=q[l]+prior[l]
        
    max=q[0]
    y_predictedvalue=0
    for m in range(26):
        if((q[m]>max)):
            max=q[m]
            y_predictedvalue=m
    y_actualvalue=y_test[i]
    


    if((y_predictedvalue == y_actualvalue)):
        count_correct_prediction=count_correct_prediction+1
    
    count_total_prediction+=1
    
accuracy =  (count_correct_prediction/count_total_prediction)*100        
```
